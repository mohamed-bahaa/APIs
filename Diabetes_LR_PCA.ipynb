```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes

# Load the diabetes dataset
df = load_diabetes(as_frame=True)
df
```

```python

# Create a DataFrame from the data
df = df.frame
df
```

```python
df.isnull().sum()
```

```python
df.corr()
```

```python
# Calculate the correlation of each feature with the target
correlations_with_target = df.corr()['target'].sort_values(ascending=False)

# Print the correlations
print(correlations_with_target)
```

```python
x=df.drop('target',axis=1)
y=df['target']
```

```python
y= y/y.max()
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_test
```

```python
from sklearn.linear_model import LinearRegression
model_1 = LinearRegression()
model_1.fit(x_train, y_train)
```

```python
# Print the coefficients
print("Coefficients:", model_1.coef_)

# Print the intercept
print("Intercept:", model_1.intercept_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.metrics import mean_squared_error

# Predict on the training set
y_train_pred = model_1.predict(x_train)

# Calculate MSE for the training set
mse_train = mean_squared_error(y_train, y_train_pred)

# Calculate MSE for the test set (already calculated in a previous cell, but recalculating here for completeness)
y_test_pred = model_1.predict(x_test)
mse_test = mean_squared_error(y_test, y_test_pred)


print(f"Mean Squared Error (Train): {mse_train}")
print(f"Mean Squared Error (Test): {mse_test}")
```

```python
from sklearn.metrics import r2_score

# Predict on the training set
y_train_pred = model_1.predict(x_train)

# Calculate R2 for the training set
r2_train = r2_score(y_train, y_train_pred)

# Predict on the test set
y_test_pred = model_1.predict(x_test)

# Calculate R2 for the test set
r2_test = r2_score(y_test, y_test_pred)

print(f"R-squared (Train): {r2_train}")
print(f"R-squared (Test): {r2_test}")
```

```python
from sklearn.decomposition import PCA

# Apply PCA to reduce features to 5
pca = PCA(n_components=5)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python
# Convert x_train_pca to a DataFrame for easier correlation calculation and display
x_train_pca_df = pd.DataFrame(x_train_pca)

# Calculate and print the correlation matrix
print("Correlation Matrix of PCA-transformed training data:")
display(x_train_pca_df.corr())
```

```python
# Train a Linear Regression model on the PCA-transformed training data
model_pca = LinearRegression()
model_pca.fit(x_train_pca, y_train)

# Predict on the PCA-transformed test set
y_test_pred_pca = model_pca.predict(x_test_pca)

# Evaluate the model using Mean Squared Error and R-squared
mse_pca = mean_squared_error(y_test, y_test_pred_pca)
r2_pca = r2_score(y_test, y_test_pred_pca)

print(f"Mean Squared Error (Linear Regression with PCA): {mse_pca}")
print(f"R-squared (Linear Regression with PCA): {r2_pca}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

# Create a pipeline with PCA and Linear Regression
pipe = Pipeline([('pca', PCA()), ('linearregression', LinearRegression())])

# Define the parameter grid for PCA
param_grid = {'pca__n_components': range(1, x_train.shape[1] + 1)}

# Set up Grid Search
grid_search = GridSearchCV(pipe, param_grid, cv=5, scoring='r2')

# Fit Grid Search to the training data
grid_search.fit(x_train, y_train)

# Print the best parameters and best score
print("Best parameters:", grid_search.best_params_)
print("Best cross-validation R-squared score:", grid_search.best_score_)
```
