```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes

# Load the diabetes dataset
df = load_diabetes(as_frame=True)
df
```

```python

# Create a DataFrame from the data
df = df.frame
df
```

```python
df.isnull().sum()
```

```python
df.corr()
```

```python
# Calculate the correlation of each feature with the target
correlations_with_target = df.corr()['target'].sort_values(ascending=False)

# Print the correlations
print(correlations_with_target)
```

```python
x=df.drop('target',axis=1)
y=df['target']
```

```python
y= y/y.max()
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_test
```

```python
from sklearn.linear_model import LinearRegression
model_1 = LinearRegression()
model_1.fit(x_train, y_train)
```

```python
# Print the coefficients
print("Coefficients:", model_1.coef_)

# Print the intercept
print("Intercept:", model_1.intercept_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.metrics import mean_squared_error

# Predict on the training set
y_train_pred = model_1.predict(x_train)

# Calculate MSE for the training set
mse_train = mean_squared_error(y_train, y_train_pred)

# Calculate MSE for the test set (already calculated in a previous cell, but recalculating here for completeness)
y_test_pred = model_1.predict(x_test)
mse_test = mean_squared_error(y_test, y_test_pred)


print(f"Mean Squared Error (Train): {mse_train}")
print(f"Mean Squared Error (Test): {mse_test}")
```

```python
from sklearn.metrics import r2_score

# Predict on the training set
y_train_pred = model_1.predict(x_train)

# Calculate R2 for the training set
r2_train = r2_score(y_train, y_train_pred)

# Predict on the test set
y_test_pred = model_1.predict(x_test)

# Calculate R2 for the test set
r2_test = r2_score(y_test, y_test_pred)

print(f"R-squared (Train): {r2_train}")
print(f"R-squared (Test): {r2_test}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

# Create a Linear Regression model
model_poly = LinearRegression()

# Train the model using the polynomial training data
model_poly.fit(x_train_poly, y_train)

# Make predictions on the polynomial test data
y_pred_poly = model_poly.predict(x_test_poly)

# Evaluate the model
mse_poly = mean_squared_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)

print(f"Mean Squared Error (Polynomial Regression): {mse_poly}")
print(f"R-squared (Polynomial Regression): {r2_poly}")
```

```python
# Predict on the polynomial training set
y_train_pred_poly = model_poly.predict(x_train_poly)

# Calculate R2 for the polynomial training set
r2_train_poly = r2_score(y_train, y_train_pred_poly)

print(f"R-squared (Polynomial Regression - Train): {r2_train_poly}")
```
